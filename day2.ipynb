```python
# A = rock
# B = paper
# C = scissors

# X = rock
# Y = paper
# Z = scissors

# total score = sum of scores each round

# score each round = shape selected (rock = 1, paper = 2, scissors = 3) +
#                    outcome of the round (lose = 0, draw = 3, win = 6)

# total score of following the guide (input)?
```

```python
input = open('day2_input.txt').read()
```

```python
# part1
score = 0
for n in input.split('\n'):
    o, p = n.split()
    #print(o)
    #print(p)
    
    if p == "X":
        if o == "C":
            score += 6
        elif o == "A":
            score += 3
        score += 1
    
    elif p == "Y":
        if o == "A":
            score += 6
        elif o == "B":
            score += 3
        score += 2
        
    elif p == "Z":
        if o == "B":
            score += 6
        elif o == "C":
            score += 3
        score += 3
    
    print(score)
```

```python
# part2
score = 0
for n in input.split('\n'):
    o, p = n.split()

    if p == "X":
        if o == "A":
            score += 3
        elif o == "B":
            score += 1
        elif o == "C":
            score += 2
        
        score += 0

    elif p == "Y":
        if o == "A":
            score += 1
        elif o == "B":
            score += 2
        elif o == "C":
            score += 3
        score += 3

    elif p =="Z":
        if o == "A":
            score += 2
        elif o == "B":
            score += 3
        elif o == "C":
            score += 1
        score += 6

    print(score)
```
